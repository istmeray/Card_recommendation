```python


# 카테고리 컬럼 리스트
category_cols_kor = [
    "보험/병원", "사무/서적/학원", "여행/레저/문화", "요식업",
    "용역/수리/건축자재", "유통", "자동차", "식료품"
]

# 🔸 숫자형 변환 (에러 방지)
for col in category_cols_kor:
    merged2[col] = pd.to_numeric(merged2[col], errors="coerce")  # 문자열 → 숫자 (오류 무시)
merged2[category_cols_kor] = merged2[category_cols_kor].fillna(0.0)

# 🔸 top2 추출 함수
def get_top2(row):
    numeric_values = row[category_cols_kor].astype(float)
    top2 = numeric_values.nlargest(2)
    cats = top2.index.tolist()
    vals = (top2.values * 100).round(1)
    return pd.Series({
        "top1": cats[0],
        "top1_pct": vals[0],
        "top2": cats[1],
        "top2_pct": vals[1],
        "top_summary": f"{cats[0]} {vals[0]}%, {cats[1]} {vals[1]}%"
    })

# 🔸 함수 적용
top_cols = merged2.apply(get_top2, axis=1)

# 🔸 결합
merged2 = pd.concat([merged2, top_cols], axis=1)

# 🔸 결과 예시 출력
print(merged2[["top1", "top1_pct", "top2", "top2_pct", "top_summary"]].head())
```
