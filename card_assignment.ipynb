```python
import random
import pandas as pd

# 카드 목록
card_list = card_benefits_df['card_name'].dropna().unique().tolist()

# 클러스터별 추천 카드 설정
cluster_card_weights = {}
for _, row in weighted_recommend_df.iterrows():
    cluster = row['cluster']
    cluster_card_weights[cluster] = [row['추천카드1'], row['추천카드2'], row['추천카드3']]

# 고객별 가중치 기반 추천 (한 행에 3개 컬럼으로)
recommendations = []

for customer_id, row in customer_df.iterrows():  # 인덱스가 고객 ID
    cluster = row['cluster']

    preferred_cards = cluster_card_weights.get(cluster, [])
    other_cards = list(set(card_list) - set(preferred_cards))

    # 추천 카드 풀: 클러스터 추천 카드는 5배 가중치
    weighted_pool = preferred_cards * 5 + other_cards
    num_cards = random.randint(1, 3)
    sampled_cards = random.sample(weighted_pool, num_cards)

    # 추천 카드 1~3개를 각각의 컬럼에 저장
    rec = {
        'customer_id': customer_id,
        '추천카드1': sampled_cards[0] if len(sampled_cards) > 0 else None,
        '추천카드2': sampled_cards[1] if len(sampled_cards) > 1 else None,
        '추천카드3': sampled_cards[2] if len(sampled_cards) > 2 else None,
    }
    recommendations.append(rec)

# 최종 결과 DataFrame
recommend_df_wide = pd.DataFrame(recommendations)
```
